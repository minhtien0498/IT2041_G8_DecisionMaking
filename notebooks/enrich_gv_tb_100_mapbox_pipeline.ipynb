# Enrich 100 BĐS Gò Vấp - Tân Bình bằng Mapbox

Notebook này chạy pipeline enrich POI bằng Mapbox cho bộ dữ liệu sạch 100 BĐS.

Input:
- `data/go_vap_tan_binh_100.json`

Output:
- `data/mapbox/go_vap_tan_binh_100_enriched_mapbox_api.json`
- `data/mapbox/go_vap_tan_binh_100_enriched_mapbox_api_checkpoint.json`
- `data/mapbox/go_vap_tan_binh_100_mapbox_errors.json`

Yêu cầu: cần có `MAPBOX_TOKEN` trong file `.env` hoặc biến môi trường.

```python
from pathlib import Path

ROOT = Path.cwd()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

input_file = ROOT / "data" / "go_vap_tan_binh_100.json"
output_file = ROOT / "data" / "mapbox" / "go_vap_tan_binh_100_enriched_mapbox_api.json"
checkpoint_file = ROOT / "data" / "mapbox" / "go_vap_tan_binh_100_enriched_mapbox_api_checkpoint.json"
error_file = ROOT / "data" / "mapbox" / "go_vap_tan_binh_100_mapbox_errors.json"

print("Input exists:", input_file.exists(), input_file)
print("Output:", output_file)
print("Checkpoint:", checkpoint_file)
print("Error log:", error_file)
```
